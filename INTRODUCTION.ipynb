#**<font color='brown'>· <u>Introduction</u>.</font>**
Social media has gradually come to occupy an ever-increasing space in many areas of our lives, leading us to conduct a significant proportion of our interactions in this virtual world.
This virtualisation of social relationships allows us to hide behind an identity that is different from who we really are. <b>Anonymity</b> allows the vanity within us to surface, for example, by presenting ourselves as much more than we are without any major consequences. However, it also facilitates the emergence of other, far more undesirable and harmful behaviours, such as harassment (cyberbullying).


The concealment of identity that we have mentioned can lead to a dangerous situation: anonymity leaves the victim defenceless, which in turn leads to silence and the problem being swept under the carpet, as well as impunity for the perpetrator. The ideal aim would be to identify the points within the network where the perpetrators are located in this environment that renders the victim powerless.

This apparent dead end may find a possible solution in the role played by emotions and their analysis.
Despite the victim’s efforts to conceal their situation from those around them, their emotions partly reflect and reveal their personal circumstances.

Often, when this sad situation comes to light, we find that those around them recall, in hindsight, that the only thing they noticed were the victim’s emotional changes.
That is why this will be one of the key elements we will incorporate into our approach to the problem.

<br>

<b>Problem to be solved</b>.

Solving the problem we have just outlined will be the central theme of the thesis we are about to develop, and can be summarised as follows:
<i>Given a social network where individuals interact with one another, and where some of them engage in harassing others whilst remaining anonymous, the aim is to be able to detect their presence and pinpoint their position within the network.
Their actions remain hidden, but their emotions are reflected in the state of each node. Using these as our sole source of information, we must be able to find the solution.

If we take into account certain studies on the contagion effect of emotions, we can exploit the way in which emotions can spread their effects locally to mathematically model the functioning of a social network. This is the approach we will follow.</i>

From a mathematical perspective, the most straightforward way to represent a social network is through a graph: a set of nodes connected by edges (<i>see fig. 1</i> ). And in order to locate the nodes under the conditions we have outlined, there are techniques relating to the dissemination of information within a network that could be applied. This is, however, a problem which, when approached algorithmically, can become computationally intractable when the network is sufficiently large.

Our aim will be to propose <b>methodological innovations</b> that are computationally feasible and that enable us to solve the problem we have set out. The existence of emotionally labelled networks, and the growing application of sentiment analysis techniques on social media, can facilitate the collection of this type of information in practice. We will carry out the study using simulation, with the simplest type of network, in order to subsequently assess the possible extension to the more general case.

<br>

<b>Research approach</b>.
1) Firstly, we will generate a simulation that reflects how a social network operates. The initial network will be a regular network (grid), and we will use formalised ways of interaction between individuals, such as those proposed by <b>Game Theory</b>, specifically the “***Iterated Prisoner’s Dilemma***” (<i>see Fig. 2</i>). This is a game in which uncertainty about the other player’s intentions plays a key role, and which is simple enough to be manageable, yet complex enough to reflect the most significant situations we might encounter (cooperating, confronting, betraying, etc., and the associated emotions).

2) The resolution phase will rely on modern artificial intelligence techniques. Specifically, we will use neural networks to process the data generated during the simulation phase, with the aim of detecting emotional patterns and regularities characteristic of how online aggressors behave.
